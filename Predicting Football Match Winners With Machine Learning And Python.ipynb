## CLEANING DATA FOR MACHINE LEARNING

```python
df.dtypes

#date             object
```

## CREATING PREDICTORS FOR ML

```python
#converting venue (away/home) column values into categorical values (astype('category')) and then into numeric values (cat.codes)

df['venue_code'] = df['venue'].astype('category').cat.codes
```

```python
df['venue_code']
# venue 'away' = 0; venue 'home' = 1
```

```python
df
```

```python
# converting opponent column values into numeric values

df['opponent_code'] = df['opponent'].astype('category').cat.codes
df
```

```python
#removing minutes from time column and converting values into integers
df['hour'] = df['time'].str.replace(":.+", "", regex = True).astype('int')
df
```

```python
#extracting day code from 'date' column
df['day_code'] = df['date'].dt.dayofweek
df
```

```python
# cponverting 'W' values into True/False and convert result values W == 1, L/D == 0
df['target'] = (df['result'] == 'W').astype('int')
```

```python
df[['result', 'target']]
```

## CREATING INITIAL MACHINE LEARNING MODEL

```python

from sklearn.ensemble import RandomForestClassifier
```

```python
#instantiate the RandomForestClassifier() and assign parameters
rf = RandomForestClassifier(n_estimators = 50, min_samples_split = 10, random_state = 1)
```

```python
#split data into train and test
train_data = df[df['date'] < '2022-01-01']
test_data = df[df['date'] > '2022-01-01']
```

```python
#creating a list of predictors 
predictors = ['venue_code', 'opponent_code', 'hour', 'day_code' ]
```

```python
# fitting data into 

rf.fit(train_data[predictors], train_data['target'])
```

```python
predictions = rf.predict(test_data[predictors])
```

```python
# evaluating if the model' predictions are accurate

from sklearn.metrics import accuracy_score
```

```python
#checking how many times we predicted correct

accuracy = accuracy_score(test_data['target'], predictions)
```

```python
accuracy
```

```python
# creating a combined dataframe to compare actual and predicted values

combined_df = pd.DataFrame(dict(actual=test_data['target'], prediction= predictions))
```

```python
pd.crosstab(index = combined_df['actual'], columns = combined_df['prediction'])
```

## IMPROVING PRECISION WITH ROLLING AVERAGE

```python
#preparing data to compute rolling averages; group data by teams

grouped_matches = df.groupby("team")

```

```python
#investigating a single group data

group = grouped_matches.get_group("Manchester City")
```

```python
group
```

```python
# create a function

def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols]  = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```

```python
#define cols varable and new_cols
cols = ["gf","ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
#group = Manchester City as defined above
rolling_averages(group, cols, new_cols)
```

```python
# apply rolling_averages to all teams

matches_rolling = df.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))

matches_rolling
```

```python
#dropping team as indexes

matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
# assigning unique index
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

## RETRAINING MACHINE LEARNING MODEL

```python
#create a function to avoid repeating steps

def make_predictions(data, predictors):
    #split data into train and test
    train_data = data[data['date'] < '2022-01-01']
    test_data = data[data['date'] > '2022-01-01']
    # fit data into RandomForestClassifier
    rf.fit(train_data[predictors], train_data['target'])
    predictions = rf.predict(test_data[predictors])
    combined_df = pd.DataFrame(dict(actual=test_data['target'], prediction= predictions))
    precision = precision_score(test_data['target'], predictions)
    return combined_df, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
precision
```

```python
#add more data to combined dataset
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
```

## COMBINING HOME AND AWAY PREDICTIONS

```python
#creating consistency across names of teams; e.x. Wolverhampton Wanderers = Wolves

class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```

```python
mapping["West Ham United"]
mapping["Arsenal"]
```

```python
#adding new column

combined["new team"] = combined["team"].map(mapping)
```

```python
combined
```

```python
#merge combined dataset to itself

merged = combined.merge(combined, left_on=["date", "new team"], right_on=["date", "opponent"])
```
